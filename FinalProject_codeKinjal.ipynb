```python
#File to load
salaries_df=pd.read_csv("C:/Users/prana/Downloads/ds_salaries.csv")
salaries_df.head()
```

```python
#Get unique values of job titles
job_titles=salaries_df["job_title"].unique()
print(job_titles)
```

```python
#Count the number of total Job Titles 
len(job_titles)
```

```python
#Calculate the total number of jobs per job title
#from the outcome below it appears that top four job titles held by number of counts is Data Scientist (143), Data Engineer (132),Data Analyst (97) and Machine learning engineer(41)

job_counts=salaries_df.groupby(['job_title']).size()
display(job_counts)
```

```python
# make groups by job titles
jt=salaries_df.groupby('job_title')
```

```python
#get all the data for Data Scientist job title
data_scientist=jt.get_group('Data Scientist')
display(data_scientist)
```

```python
#print the mean of salary in usd for data scientist
print(data_scientist['salary_in_usd'].mean())
```
